# Heat-Related Mortality Figures

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "/Users/jessesmac/Myprojects/data/coiled_output/HRD_CalculationsTable_ByYear_2013-2022_OT_SMT.csv"
df = pd.read_csv(file_path)

# Extract columns
years = df['Year']
otm_hrd = df['OTM HRD']
ehm_hrd = df['EHM HRD']
official_deaths = df['Official Deaths']

# Determine x-tick labels for every other year
xtick_labels = [str(year) if i % 2 == 0 else '' for i, year in enumerate(years)]

# Create the plot
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # No sharey

# === Left Panel: OTM vs EHM ===
axes[0].plot(years, otm_hrd, marker='o', markersize=8, linestyle='-.', color='#FFA726', linewidth=2.5, label='OTM HRD')
axes[0].plot(years, ehm_hrd, marker='s', markersize=8, linestyle='--', color='#E57373', linewidth=2.5, label='XHM HRD')
axes[0].set_title('OTM HRD vs XHM HRD', fontsize=22)
axes[0].set_ylabel('Number of Deaths', fontsize=20)
axes[0].legend(fontsize=16)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)
axes[0].tick_params(axis='both', labelsize=18)
axes[0].set_xticks(years)
axes[0].set_xticklabels(xtick_labels, rotation=45)

# === Right Panel: EHM vs Official ===
axes[1].plot(years, ehm_hrd, marker='s', markersize=8, linestyle='--', color='#E57373', linewidth=2.5, label='XHM HRD')
axes[1].plot(years, official_deaths, marker='d', markersize=8, linestyle='-', color='#64B5F6', linewidth=2.5, label='Official Deaths')
axes[1].set_title('XHM HRD vs Official Deaths', fontsize=22)
axes[1].legend(fontsize=16)
axes[1].grid(axis='y', linestyle='--', alpha=0.6)
axes[1].tick_params(axis='both', labelsize=18)
axes[1].set_xticks(years)
axes[1].set_xticklabels(xtick_labels, rotation=45)

# General Layout
plt.suptitle('Comparison of Modeled and Official HRD', fontsize=22)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Load Excess HRD File ===
excess_path = "/Users/jessesmac/Myprojects/data/coiled_output/Excess_HRD.csv"
excess_df = pd.read_csv(excess_path)

# === Extract values ===
years = excess_df['Year']
excess_1950s = excess_df['Excess_HRD_vs_1950s']
excess_1990s = excess_df['Excess_HRD_vs_1990s']

# === Setup Plot ===
plt.figure(figsize=(10, 6))

plt.plot(years, excess_1950s, marker='D', markersize=8, linestyle='-', 
         color='#4FC3F7', linewidth=2.5, label="1950s baseline")
plt.plot(years, excess_1990s, marker='o', markersize=8, linestyle='-', 
         color='#81C784', linewidth=2.5, label="1990s baseline")
plt.axhline(0, color='black', linestyle='--', linewidth=1.2)

# === Formatting ===
plt.title('Excess Heat Related Deaths', fontsize=22)
plt.ylabel('Excess Deaths', fontsize=20)
plt.xticks(years, rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16, loc='upper center')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# === Show Plot ===
plt.show()
```
